### Machine Learning

```python
weather_df['tomorrow_max_temp'] = weather_df.shift(-1)["tmax"]
```

```python
weather_df
```

```python
weather_df = weather_df.ffill()
weather_df
```

```python
weather_df.corr(numeric_only=True)
```

```python
from sklearn.linear_model import Ridge

ridge_model = Ridge(alpha=.1)

predictors = weather_df.columns[~weather_df.columns.isin(["tomorrow_max_temp", "name", "station"])]
predictors
```

```python
def backtesting(weather_df, ridge_model, predictors, start = 3650, step=90):
    all_predictions = []

    for i in range(start, weather_df.shape[0],90):
        train = weather_df.iloc[:i, :]
        test = weather_df.iloc[i:(i+step), :]

        ridge_model.fit(train[predictors], train["tomorrow_max_temp"])

        predictions = ridge_model.predict(test[predictors])

        predictions = pd.Series(predictions, index=test.index)
        combined = pd.concat([test["tomorrow_max_temp"], predictions], axis=1)

        combined.columns = ["actual", "predictions"]
        combined["diff"] = (combined["predictions"]-combined["actual"]).abs()

        all_predictions.append(combined)
    
    return pd.concat(all_predictions)
```

```python
predictions = backtesting(weather_df, ridge_model, predictors)
predictions
```

```python
from sklearn.metrics import mean_absolute_error

mean_absolute_error(predictions["actual"], predictions["predictions"])
```

```python
def pct_diff(old, new):
    return (new-old)/old

def rolling_average(weather, horizon, col):
    label = f"rolling_{horizon}_{col}"
    
    weather[label] = weather[col].rolling(horizon).mean()
    weather[f"{label}_pct"] = pct_diff(weather[label], weather[col])
    return weather

rolling_horizons = [3,14]

for horizon in rolling_horizons:
    for col in ["tmax", "tmin", "prcp"]:
        weather_df = rolling_average(weather_df, horizon, col)
```

```python
weather_df
```

```python
weather_df = weather_df.iloc[14:, :]
weather_df
```

```python
weather_df = weather_df.fillna(0)
weather_df
```

```python
def expand_mean(df):
    return df.expanding(1).mean()

for col in ["tmax", "tmin", "prcp"]:
    weather_df[f"month_avg_{col}"] = weather_df[col].groupby(weather_df.index.month, group_keys=False).apply(expand_mean)
    weather_df[f"day_avg_{col}"] = weather_df[col].groupby(weather_df.index.day_of_year, group_keys=False).apply(expand_mean)

weather_df
```

```python
predictors = weather_df.columns[~weather_df.columns.isin(["tomorrow_max_temp", "name", "station"])]
predictors
```

```python
predictions = backtesting(weather_df, ridge_model, predictors)
predictions
```

```python
mean_absolute_error(predictions["actual"], predictions["predictions"])
```

```python
from sklearn.ensemble import RandomForestRegressor

random_forest_model = RandomForestRegressor(max_depth=8, random_state=0)

def backtesting(weather_df, model, predictors, start = 3650, step=90):
    all_predictions = []

    for i in range(start, weather_df.shape[0],90):
        train = weather_df.iloc[:i, :]
        test = weather_df.iloc[i:(i+step), :]

        model.fit(train[predictors], train["tomorrow_max_temp"])

        predictions = model.predict(test[predictors])

        predictions = pd.Series(predictions, index=test.index)
        combined = pd.concat([test["tomorrow_max_temp"], predictions], axis=1)

        combined.columns = ["actual", "predictions"]
        combined["diff"] = (combined["predictions"]-combined["actual"]).abs()

        all_predictions.append(combined)
    
    return pd.concat(all_predictions)


predictions_new = backtesting(weather_df,random_forest_model, predictors)
predictions_new
```
